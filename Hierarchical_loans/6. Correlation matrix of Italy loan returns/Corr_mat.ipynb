```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(context='paper', style='white')
```

```python
df = pd.read_parquet('./loans_full_hierarchy.parquet')
df['date'] = df.index
```

```python
plt.figure(figsize=(15,10))
corr = df.drop(columns=['date']).pct_change().dropna().corr('spearman').iloc[:26,:26]
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask = mask, cmap='coolwarm', vmin=-1, annot = True, vmax = 1)
plt.savefig('./corr_mat_regions.png', transparent = True, dpi = 400)
```
